<b>1. WAP to demonstrate effects of changing values of brightness and contrast.</b>

```python
import numpy as np
from google.colab.patches import cv2_imshow
import cv2
image = cv2.imread("/content/image.jpeg")

new_image = np.zeros(image.shape, image.dtype)
alpha = 1.0 # Simple contrast control
beta = 0    # Simple brightness control
alpha = float(input('* Enter the alpha value [1.0-3.0]: '))
beta = int(input('* Enter the beta value [0-100]: '))
for y in range(image.shape[0]):
    for x in range(image.shape[1]):
        for c in range(image.shape[2]):
            new_image[y,x,c] = np.clip(alpha*image[y,x,c] + beta, 0, 255)

cv2_imshow(image)
cv2_imshow(new_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

<b> 2. WAP to show effect of inversion on the image </b>

```python
import numpy as np
from google.colab.patches import cv2_imshow
import cv2
img = cv2.imread("/content/image.jpeg")
cv2_imshow(img)
img_not = cv2.bitwise_not(img)
cv2_imshow(img_not)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

<b> 3. WAP to transform image to an histogram. </b>

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('/content/image.jpeg', -1)
cv2_imshow(img)

color = ('b','g','r')
for channel,col in enumerate(color):
    histr = cv2.calcHist([img],[channel],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])
plt.title('Histogram for color scale picture')
plt.show()
cv2.waitKey(0)         
cv2.destroyAllWindows()
```

<b>4. WAP to implement the process of histogram stretching.</b>

```python
import cv2
import numpy as np
 
img1 = cv2.imread('/content/image.jpeg',0)
# print(img1)
# Create zeros array to store the stretched image
minmax_img = np.zeros((img1.shape[0],img1.shape[1]),dtype = 'uint8')
 
for i in range(img1.shape[0]):
    for j in range(img1.shape[1]):
        minmax_img[i,j] = 255*(img1[i,j]-np.min(img1))/(np.max(img1)-np.min(img1))
 
cv2_imshow(minmax_img)
cv2.waitKey(0)
```
